```python
healthindexdata = {}
for i, area in enumerate([areas.loc[10], areas.loc[40], areas.loc[100], areas.loc[150], areas.loc[200], areas.loc[280], areas.loc[220], areas.loc[110], areas.loc[300]]):
    print(i, 'area: ', area)

    data = {}

    for domain in domains:

        data[domain] = {}
        data[domain]['total'] = scores(domain_data, area, domain)
        data[domain]['subdomains'] = {}

        for subdomain in get_subdomains(domain):

            data[domain]['subdomains'][subdomain] = {}
            data[domain]['subdomains'][subdomain]['total'] = scores(subdomain_data, area, subdomain)
            # data[domain]['subdomains'][subdomain]['indicators'] = {}

            # for indicator in get_indicators(subdomain):

            #     data[domain]['subdomains'][subdomain]['indicators'][indicator] = scores(indicator_data, area, indicator)

    # OVERALL

    data['Overall'] = {}
    data['Overall']['total'] = {}

    for year in years:
        data['Overall']['total'][year] = {}
        data['Overall']['total'][year]['value'] = round(topindex[topindex['AREANM'] == area].iloc[0][yearLU[year]], 1)
        data['Overall']['total'][year]['rank'] = list(topindex.sort_values(by=yearLU[year], ascending=False)['AREANM']).index(area)+1

        # CHANGE1YEAR
        if year == 2015:
            data[domain]['total'][year]['Change1year'] = 'NaN'
            data[domain]['total'][year]['Change1year Rank'] = 'NaN'
        else:
            change1 = topindex[topindex['AREANM'] == area].iloc[0][yearLU[year]] - topindex[topindex['AREANM'] == area].iloc[0][yearLU[year-1]]
            data[domain]['total'][year]['Change1year'] = round(change1, 1)

            change1list = sorted([topindex.loc[i][yearLU[year]] - topindex.loc[i][yearLU[year-1]] for i in topindex.index], reverse=True)
            data[domain]['total'][year]['Change1year Rank'] = change1list.index(change1) + 1

        # CHANGE4YEAR
        if year == 2019:
            change4 = topindex[topindex['AREANM'] == area].iloc[0][yearLU[2019]] - topindex[topindex['AREANM'] == area].iloc[0][yearLU[2015]]
            data[domain]['total'][year]['Change4year'] = round(change4, 1)

            change4list = sorted([topindex.loc[i][yearLU[2019]] - topindex.loc[i][yearLU[2015]] for i in topindex.index], reverse=True)
            data[domain]['total'][year]['Change4year Rank'] = change4list.index(change4) + 1
        else:
            data[domain]['total'][year]['Change4year'] = 'NaN'
    #         data[domain]['total'][year]['Change4year Rank'] = 'NaN'


    subds = {**data['Healthy People']['subdomains'], **data['Healthy Lives']['subdomains'], **data['Healthy Places']['subdomains']}
    stories = sorted([{**{'metric': i}, **subds[i]['total'][2019]} for i in subds], reverse=False, key= lambda x: storyorder(x))

    if max([storyorderneg(i) for i in stories][:4]) < 0:
        camstories1 = stories
        print('None of top four are positive')
        print(stories)
        print(' . *   *   * ')
        sneg = [storyorderneg(i) for i in stories]
        vpos = next(x for x in sneg if x > 0)
        fpos = sneg.index(vpos)
        stories.insert(3, stories.pop(fpos))
        print(stories)
        camstories2 = stories


    place = {}
    place["area"] = [topindex['AREACD'].loc[i] for i in topindex.index if topindex['AREANM'].loc[i] == area][0]
    place["name"] = area
    place["data"] = data
    place['stories'] = stories
    place['region'] = regLU[place["area"]]

    healthindexdata[place['area']] = place
```

```python
def storyorderneg(x):
    if x['metric'] == 'Access to green space':
        return x['rank']
    else:
        if min(abs(x['rank']), abs(x['Change1year Rank']), abs(x['Change4year Rank'])) == abs(x['rank']):
            return x['rank']
        elif min(abs(x['rank']), abs(x['Change1year Rank']), abs(x['Change4year Rank'])) == abs(x['Change1year Rank']):
            return x['Change1year Rank']
        elif min(abs(x['rank']), abs(x['Change1year Rank']), abs(x['Change4year Rank'])) == abs(x['Change4year Rank']):
            return x['Change4year Rank']
```

```python
def storyorder(x):
    if x['metric'] == 'Access to green space':
        return abs(x['rank'])
    else:
        return min(abs(x['rank']), abs(x['Change1year Rank']), abs(x['Change4year Rank']))
stories
```

```python
def scores(df, area, measure):
    df_filt = df[(df['DATANM'] == measure)]
    region = regLU[df[df['AREANM'] == area].iloc[0]['AREACD']]
    df_filt_reg = df_filt[df_filt['RGN21CD'] == region['code']]
    df_reg = df[df['RGN21CD'] == region['code']]

    tot_num = 307
    reg_num = len(df_filt_reg)

    tempob = {}
    for year in years:
        # CHANGE1YEAR
        if year == 2015:
            Change1year = 'NaN'
            Change1yearRank = 'NaN'
            Change1yearRankReg = 'NaN'
        else:
            change1 = df_filt[df_filt['AREANM'] == area].iloc[0][yearLU[year]] - df_filt[df_filt['AREANM'] == area].iloc[0][yearLU[year-1]]
            Change1year = round(change1, 1)

            change1list = sorted([df_filt.loc[i][yearLU[year]] - df_filt.loc[i][yearLU[year-1]] for i in df_filt.index], reverse=True)
            change1listReg = sorted([df_filt_reg.loc[i][yearLU[year]] - df_filt_reg.loc[i][yearLU[year-1]] for i in df_filt_reg.index], reverse=True)

            Change1yearRank = change1list.index(change1) + 1
            if Change1yearRank>(tot_num/2):
                Change1yearRank = (Change1yearRank-tot_num)-1

            Change1yearRankReg = change1listReg.index(change1) + 1
            if Change1yearRankReg>(reg_num/2):
                Change1yearRankReg = (Change1yearRankReg-reg_num)-1

        # CHANGE4YEAR
        if year == 2019:
            change4 = df_filt[df_filt['AREANM'] == area].iloc[0][yearLU[2019]] - df_filt[df_filt['AREANM'] == area].iloc[0][yearLU[2015]]
            Change4year = round(change4, 1)

            change4list = sorted([df_filt.loc[i][yearLU[2019]] - df_filt.loc[i][yearLU[2015]] for i in df_filt.index], reverse=True)
            change4listReg = sorted([df_filt_reg.loc[i][yearLU[2019]] - df_filt_reg.loc[i][yearLU[2015]] for i in df_filt_reg.index], reverse=True)

            Change4yearRank = change4list.index(change4) + 1
            if Change4yearRank>(tot_num/2):
                Change4yearRank = (Change4yearRank-tot_num)-1

            Change4yearRankReg = change4listReg.index(change4) + 1
            if Change4yearRankReg>(reg_num/2):
                Change4yearRankReg = (Change4yearRankReg-reg_num)-1


        else:
            Change4year = 'NaN'
            Change4yearRank = 'NaN'
            Change4yearRankReg = 'NaN'

        rank = list(df_filt.sort_values(by=yearLU[year], ascending=False)['AREANM']).index(area)+1
        if rank>(tot_num/2):
            rank = (rank-tot_num)-1

        rankReg = list(df_filt_reg.sort_values(by=yearLU[year], ascending=False)['AREANM']).index(area)+1
        if rankReg>(reg_num/2):
            rankReg = (rankReg-reg_num)-1

        tempob[year] = {
            'value': round(df_filt[df_filt['AREANM'] == area].iloc[0][yearLU[year]], 1),
            'rank': rank,
            'rank_Reg': rankReg,
            'Change1year': Change1year,
            'Change1year Rank': Change1yearRank,
            'Change1year Rank_Reg': Change1yearRankReg,
            'Change4year': Change4year,
            'Change4year Rank': Change4yearRank,
            'Change4year Rank_Reg': Change4yearRankReg
        }
    return tempob
scores(subdomain_data, areas[0], 'Access to green space')
```
